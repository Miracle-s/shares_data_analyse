- date：日期
- open：开盘价
- high：最高价
- close：收盘价
- low：最低价
- volume：成交量
- price_change：价格变动
- p_change：涨跌幅
- ma5：5日均价
- ma10：10日均价
- ma20:20日均价
- v_ma5:5日均量
- v_ma10:10日均量
- v_ma20:20日均量
- turnover:换手率[注：指数无此项]

```python
import tushare as ts
# 获取 上证50ETF 510050 （上证大蓝筹）
# 获取 沪深300ETF 510300 (沪深大蓝筹)
# 获取 创业板ETF 159952 （创业板整体行情）
data_50 = ts.get_hist_data('510050', start='2010-01-01', end='2020-01-01')
data_300 = ts.get_hist_data('510300', start='2010-01-01', end='2020-01-01')
data_cy = ts.get_hist_data('159952', start='2010-01-01', end='2020-01-01')

column_dict = {'date':'日期','open':'开盘价','high':'最高价','close':'收盘价','low':'最低价','volume':'成交量',\
'price_change':'价格变动','p_change':'涨跌幅','ma5':'5日均价','ma10':'10日均价','ma20':'20日均价','v_ma5':'5日均量',\
'v_ma10':'10日均量','v_ma20':'20日均量','turnover':'换手率'}

data_50.rename(columns=column_dict, inplace=True)
data_300.rename(columns=column_dict, inplace=True)
data_cy.rename(columns=column_dict, inplace=True)
data_50
```
